## **Data Cleaning**

**1. Remove Duplicates**

```python
df.info()
```

```python
df.shape
```

```python
df = df.drop_duplicates()
```

```python
df.shape
```

**2. Changing column name**

```python
df.columns
```

```python
df = df.rename(columns = {'Invoice ID' : 'id', 'Customer type' : 'cust_type', 
                     'Product line' : 'product', 'gross margin percentage' : 'gross_margin_perc', 
                     'Unit price' : 'unit_price'})
df
```

**2. Changing ID column and it's datatype**

```python
df
```

```python
df['id'] = df['id'].str.replace('-', '')
df['Date'] = df['Date'].str.replace('/', '-')
```

```python
df = df.astype({'id' : 'int'})
```

```python
df.info()
```

```python
df
```

## **Data Visualization**

```python
purchased_quantity = df.groupby(by=['product','Gender'], as_index = False).agg({'Quantity' : 'sum'})
purchased_quantity
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,5))
sns.barplot(x = 'product', y = 'Quantity', hue = 'Gender', data = purchased_quantity, width = 0.5, palette='dark:orange')
plt.title('Gender wise Quantity of product purchased', fontdict={'fontweight' : 'bold', 'fontsize' : 20, 'color' : 'Orange'})
plt.grid(axis = 'y')

plt.show()
```

```python
purchasing_power = df.groupby(by=['product','Gender'], as_index = False).agg({'Total' : 'sum'})
purchasing_power
```

```python
plt.figure(figsize=(20,5))
sns.barplot(x = 'product', y = 'Total', hue = 'Gender', data = purchasing_power, width = 0.6, palette = 'dark:green')
plt.title('Purchasing Power of male and female', fontdict={'fontweight' : 'bold', 'fontsize' : 20, 'color' : 'Orange'})
plt.ylabel('Total price in USD')
plt.grid(axis = 'y')

plt.show()
```

```python
a = df.groupby(by=['City', 'Gender'], as_index = False).agg({'Total' : 'sum'})
a
```

```python
city_wise_purchasing_power = df.groupby(by=['City', 'Gender'], as_index = False).size()
city_wise_purchasing_power
```

```python
plt.figure(figsize = (10,5))
sns.barplot(x = 'City', y = 'size', hue = 'Gender', data = city_wise_purchasing_power, palette='dark:b', width = 0.30)
plt.grid(axis = 'y')
plt.xlabel('city', fontdict = {'fontsize' : 15})
plt.ylabel('Number of person', fontdict = {'fontsize' : 10})
plt.title('Gender wise purchasing power per City', fontdict = {'fontsize' : 20, 'color' : 'g'})
plt.show()
```

```python
product_and_gender_wise_gross_income = df.groupby(by = ['product', 'Gender'], as_index = False).agg({'gross income' : 'sum'})
product_and_gender_wise_gross_income
```

```python
plt.figure(figsize = (15,5))
sns.barplot(x = 'product', y = product_and_gender_wise_gross_income['gross income'], hue = 'Gender', 
            data = product_and_gender_wise_gross_income, palette = 'dark:blue', width = 0.3)

plt.grid(axis = 'y')
plt.legend(loc = 'upper right', title = 'Gender')

plt.xlabel('Producr name', fontdict = {'fontsize' : 10})
plt.ylabel('Total gross income in USD', fontdict = {'fontsize' : 10})
plt.title('Gender wise gross income per product', fontdict = {'fontsize' : 20, 'color' : 'g'})

plt.show()
```

```python
product_wise_gross_income = df.groupby(by = ['product'], 
                                       as_index = False).agg({'gross income' : 'sum'}).sort_values(by = ['gross income'], ascending = False)
product_wise_gross_income
```

```python
plt.figure(figsize = (15, 10))
colors = ['orange', '#e2e2e2', 'red', 'yellow', '#bb2b', 'green']
plt.pie(product_wise_gross_income['gross income'], radius = 0.8, wedgeprops = {'linewidth' : 1, 'edgecolor':'black'},
        autopct = '%.2f %%', explode = [0.1, 0, 0, 0, 0, 0], colors = colors, labels = product_wise_gross_income['product'])

plt.legend(labels = product_wise_gross_income['product'], title = 'Products')
plt.title('Product wise gross income', fontdict = {'fontsize' : 20, 'color' : 'g'})

plt.show()
```
